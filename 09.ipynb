```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Đường Dẫn 
my_path = 'C:/Users/ACER/TGMT/src_cv/img/07/1.jpg'
#Đọc Ảnh
im = Image.open(my_path)
plt.imshow(im)
```

```python
#Chuyển Sang Ảnh Xám
im_gray = im.convert('L')
plt.imshow(im_gray,cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
#Tạo Một Hình Vẽ Mới
plt.figure()
plt.hist(im_array.flatten(), bins=128)
plt.title('Biểu đồ Histogram của ảnh')
plt.xlabel('Giá trị Pixel')
plt.ylabel('Số lượng')
plt.show()
```

```python
#Làm nổi bật đường biên
plt.figure()
plt.gray()
plt.contour(im_gray, origin='image')
plt.show()
```

```python
#Đường dẫn ảnh
my_path= 'C:/Users/ACER/TGMT/src_cv/img/09/1.jpg'
#Đọc ảnh và hiển thị ảnh với trục tọa độ
im = Image.open(my_path).convert("L")
# Hiển thị ảnh trên trục tọa độ
plt.imshow(im)
# Tạo một hình vẽ mới
plt.figure()
# Đặt màu sắc là xám
plt.gray()
# Vẽ biểu đồ đường biên của ảnh
plt.contour(im, origin="image")
# Đảm bảo tỷ lệ trục x và trục y trên biểu đồ là bằng nhau
plt.axis('equal')
# Hiển thị ảnh
plt.show()
```
