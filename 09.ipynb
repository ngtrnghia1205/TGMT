```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
#ng Dn 
my_path = 'C:/Users/ACER/TGMT/src_cv/img/07/1.jpg'
#c nh
im = Image.open(my_path)
plt.imshow(im)
```

```python
#Chuyn Sang nh Xm
im_gray = im.convert('L')
plt.imshow(im_gray,cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
#To Mt Hnh V Mi
plt.figure()
plt.hist(im_array.flatten(), bins=128)
plt.title('Biu  Histogram ca nh')
plt.xlabel('Gi tr Pixel')
plt.ylabel('S lng')
plt.show()
```

```python
#Lm ni bt ng bin
plt.figure()
plt.gray()
plt.contour(im_gray, origin='image')
plt.show()
```
