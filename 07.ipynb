```python
x = [1,2,3,4,5,6]
y = [10,16,30,15,60,10]
plt.plot(x,y,color = 'red')#Màu sắc
plt.title('Ví dụ biểu đồ')#Title
plt.show()#Hiển thị
```

```python
#Tắt hệ trục tọa độ
x = [1,2,3,4,5,6]
y = [10,16,30,15,60,10]

plt.plot(x,y,color = 'red')# Thay đổi màu thành đỏ
plt.title('Biểu đồ ví dụ với màu ') # Thêm tiêu đề
#Tắt hệ trục tọa độ
plt.axis('off') # Tắt hệ trục tọa độ
plt.show()
```
