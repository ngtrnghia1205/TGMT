```python
#Thêm thư viện 
from PIL import Image
from imgtools import *
```

```python
#Đường dẫn ảnh
my_path = 'C:/Users/ACER/TGMT/src_cv/img/04/1.jpg'
```

```python
#Đọc ảnh
img = load_image(my_path)
```

```python
display(img)
```

```python
#Định vị khu vực cắt
region=(100,75,175,160)
#Cắt vùng ảnh
cropped_image = img.crop(region)
display(cropped_image)
```

```python
#Dán
#Xác định vị trị cần dán
paste_position = (100,100)
img.paste(cropped_image, paste_position)
display(img)
```
